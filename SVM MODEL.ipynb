```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples=40, centers=2, random_state=6)
```

```python
# create classifier

clf = svm.SVC(kernel='linear', C=1000)
```

```python
# fitting the model

clf.fit(X, y)
```

```python
# The below set of code is using plotting libraries from matplotlib to visually display the data points,
#classification line and the support vectors. We are done with creating a trained SVM classifier by using only the above lines. 
#The below lines are completely optional and used for viewing purposes only.
#In a production use case, you might not desire to plot the data points on a graph.
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=plt.cm.Paired)

plt.scatter(X[:, 0], X[:, 1], c=y, cmap='winter')

# plot the decision function
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# create grid to evaluate model
xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = clf.decision_function(xy).reshape(XX.shape)

# plot decision boundary and margins
ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
           linestyles=['--', '-', '--'])
# plot support vectors
ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100,
           linewidth=1, facecolors='none', edgecolors='k')
plt.show()
```
